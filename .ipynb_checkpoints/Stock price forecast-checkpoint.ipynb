# Brownian movement

\begin{equation*}
\large{Price\ today = Price\ yesterday *  e ^ {\mu - \frac{1}{2}\sigma^2 + \sigma * Z(Rand(0;1))}}
\end{equation*} 

```python
NVDA_mu = Returns['NVDA'].mean()
NVDA_mu
```

```python
NVDA_var = Returns['NVDA'].var()
NVDA_var
```

```python
NVDA_std = Returns['NVDA'].std()
NVDA_std
```

\begin{equation*}
\large{Drift = \mu - \frac{1}{2}\sigma^2}
\end{equation*} 

```python
NVDA_drift = NVDA_mu - 1/2*NVDA_var
NVDA_drift
```

```python
t_interval = 60
iterations_1 = 100
```

```python
NVDA_daily_returns = np.exp(NVDA_drift + NVDA_std * norm.ppf(np.random.rand(t_interval,iterations_1)))
NVDA_daily_returns
```

```python
NVDA_daily_returns.shape
```

```python
NVDA_S0 = data['NVDA'].iloc[-1]
NVDA_S0
```

```python
NVDA_price_list = np.zeros_like(NVDA_daily_returns)
NVDA_price_list
```

```python
NVDA_price_list[0] = NVDA_S0
NVDA_price_list
```

```python
for t in range(1,t_interval):
    NVDA_price_list[t] = NVDA_price_list[t-1] * NVDA_daily_returns[t]

NVDA_price_list
```

```python
plt.figure(figsize=(12,8))
plt.plot(NVDA_price_list)
plt.xlim(left = 0,right = t_interval-1)
plt.title('NVIDIA Corporation - Monte Carlo simulation',pad = 16,
          fontdict=
          {'fontsize': 20,
           'fontweight' : 5,
           'verticalalignment': 'baseline',
           'horizontalalignment': 'center'}
         )
plt.ylabel('Price', labelpad = 10,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.xlabel('Number of Days', labelpad = 16,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.show()
```

```python
NVDA_price_list.min(), NVDA_price_list.max()
```

```python
NVDA_pred_price = []

for i in range(1,t_interval):
    NVDA_pred_price.append(NVDA_price_list[i].mean())

NVDA_pred_price
```

```python
plt.figure(figsize=(12,8))
plt.plot(NVDA_pred_price)
plt.xlim(left = 0,right = t_interval-1)
plt.title('NVIDIA Corporation - Price forecast',pad = 16,
          fontdict=
          {'fontsize': 20,
           'fontweight' : 5,
           'verticalalignment': 'baseline',
           'horizontalalignment': 'center'}
         )
plt.ylabel('Price', labelpad = 10,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.xlabel('Number of Days', labelpad = 16,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.show()
```

## Test vs Training

```python
NVDA_training = data['NVDA'].iloc[:-60]
NVDA_test = data['NVDA'].iloc[-60:]
NVDA_test
```

```python
iterations_training = 20

NVDA_returns_training = np.exp(NVDA_drift + NVDA_std * norm.ppf(np.random.rand(t_interval,iterations_training)))

NVDA_p_list_training = np.zeros_like(NVDA_returns_training)
NVDA_p_list_training[0] = NVDA_training[-1]

for t in range(1,t_interval):
    NVDA_p_list_training[t] = NVDA_p_list_training[t-1] * NVDA_returns_training[t]
    
NVDA_Phat_training = []

for i in range(1,t_interval):
    NVDA_Phat_training.append(NVDA_p_list_training[i].mean())
```

```python
plt.figure(figsize=(12,8))
plt.plot(NVDA_Phat_training)
plt.plot(NVDA_test.values)
plt.xlim(left = 0,right = t_interval-1)
plt.title('NVIDIA Corporation - Price forecast',pad = 16,
          fontdict=
          {'fontsize': 20,
           'fontweight' : 5,
           'verticalalignment': 'baseline',
           'horizontalalignment': 'center'}
         )
plt.ylabel('Price', labelpad = 10,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.xlabel('Number of Days', labelpad = 16,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.show()
```

```python
metrics.explained_variance_score(NVDA_test.values[:59],NVDA_Phat_training)
```

```python
metrics.mean_squared_error(NVDA_test.values[:59],NVDA_Phat_training)**0.5
```
