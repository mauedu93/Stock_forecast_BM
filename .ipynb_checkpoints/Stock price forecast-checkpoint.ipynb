```python
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn import metrics

%matplotlib inline
```

```python
stock = 'INTC'
data = pd.DataFrame()
data[stock] = wb.DataReader(stock, data_source = 'yahoo', start = '2000-1-1')['Adj Close']
    
data
```

```python
Returns = np.log(data.pct_change()+1)
Returns
```

```python
Returns.plot(xlim=('2000-01-03',Returns.index.max()), figsize=(12,8))
plt.title(stock + ' Corporation - Historical returns',
          fontdict=
          {'fontsize': 20,
           'fontweight' : 5,
           'verticalalignment': 'baseline',
           'horizontalalignment': 'center'}
         )
plt.ylabel('Returns', labelpad = 10,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.xlabel('Date', labelpad = 12,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.show()
```

# Brownian movement

\begin{equation*}
\large{Price\ today = Price\ yesterday *  e ^ {\mu - \frac{1}{2}\sigma^2 + \sigma * Z(Rand(0;1))}}
\end{equation*} 

```python
Mu = Returns.mean()
Mu
```

```python
var = Returns.var()
var
```

```python
std = Returns.std()
std
```

\begin{equation*}
\large{Drift = \mu - \frac{1}{2}\sigma^2}
\end{equation*} 

```python
drift = Mu - 1/2*var
drift
```

```python
t_interval = 60
iterations_1 = 100
```

```python
daily_returns = np.exp(np.array(drift) + 
                       np.array(std) * norm.ppf(np.random.rand(t_interval,iterations_1)))
daily_returns
```

```python
daily_returns.shape
```

```python
S0 = data.iloc[-1]
S0
```

```python
price_list = np.zeros_like(daily_returns)
price_list
```

```python
price_list[0] = S0
price_list
```

```python
for t in range(1,t_interval):
    price_list[t] = price_list[t-1] * daily_returns[t]

price_list
```

```python
plt.figure(figsize=(12,8))
plt.plot(price_list)
plt.xlim(left = 0,right = t_interval-1)
plt.title(stock  + ' Corporation - Monte Carlo simulation',pad = 16,
          fontdict=
          {'fontsize': 20,
           'fontweight' : 5,
           'verticalalignment': 'baseline',
           'horizontalalignment': 'center'}
         )
plt.ylabel('Price', labelpad = 10,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.xlabel('Number of Days', labelpad = 16,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.show()
```

```python
price_list.min(), price_list.max()
```

```python
print('first quartile:',np.quantile(price_list,0.25))
print('second quartile:',np.quantile(price_list,0.5))
print('third quartile:',np.quantile(price_list,0.75))
```

```python
pred_price = []

for i in range(1,t_interval):
    pred_price.append(price_list[i].mean())

pred_price
```

```python
plt.figure(figsize=(12,8))
plt.plot(pred_price)
plt.xlim(left = 0,right = t_interval-1)
plt.title(stock + ' Corporation - Price forecast',pad = 16,
          fontdict=
          {'fontsize': 20,
           'fontweight' : 5,
           'verticalalignment': 'baseline',
           'horizontalalignment': 'center'}
         )
plt.ylabel('Price', labelpad = 10,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.xlabel('Number of Days', labelpad = 16,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.show()
```

## Test vs Training

```python
training = data[stock].iloc[:-60]
test = data[stock].iloc[-60:]
```

```python
iterations_training = 15

returns_training = np.exp(np.array(drift) 
                          + np.array(std) * norm.ppf(
                              np.random.rand(t_interval,iterations_training)))

p_list_training = np.zeros_like(returns_training)
p_list_training[0] = training[-1]

for t in range(1,t_interval):
    p_list_training[t] = p_list_training[t-1] * returns_training[t]
    
Phat_training = []

for i in range(1,t_interval):
    Phat_training.append(p_list_training[i].mean())
```

```python
plt.figure(figsize=(12,8))
plt.plot(Phat_training)
plt.plot(test.values, 'r-')
plt.xlim(left = 0,right = t_interval-1)
plt.title(stock + ' Corporation - Price forecast',pad = 16,
          fontdict=
          {'fontsize': 20,
           'fontweight' : 5,
           'verticalalignment': 'baseline',
           'horizontalalignment': 'center'}
         )
plt.ylabel('Price', labelpad = 10,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.xlabel('Number of Days', labelpad = 16,
           fontdict =
           {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'}
          )
plt.show()
```

```python
metrics.explained_variance_score(test.values[:59],Phat_training)
```

```python
RMSE = metrics.mean_squared_error(test.values[:59],Phat_training)**0.5
print('$',round(RMSE,2))
```

```python
MAE = metrics.mean_absolute_error(test.values[:59],Phat_training)
print('$',round(MAE,2))
```
